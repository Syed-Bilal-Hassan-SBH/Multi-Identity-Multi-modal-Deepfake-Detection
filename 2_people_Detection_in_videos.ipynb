# Sepearting two person videos

```python
import cv2
import os
import shutil
from tqdm import tqdm
# Path to the main folder containing subfolders of videos
main_input_folder = r'F:\DFDC\dfdc_train_part_19'  # Update this with your actual input folder path
output_folder = r'F:\DFDC\Data_2_people\Videos_19'  # Where videos with >1 person will be moved

# Load a pre-trained model (MobileNetSSD in this case)
model_path = r'C:\Users\MASTER\Desktop\Fast NU\Personal Project\Research\models'  # Path to the folder where the model files are stored
net = cv2.dnn.readNetFromCaffe(os.path.join(model_path, 'deploy.prototxt'), 
                               os.path.join(model_path, 'mobilenet_iter_73000.caffemodel'))
```

```python
# Function to detect the number of people in a frame
def count_people(frame):
    (h, w) = frame.shape[:2]
    blob = cv2.dnn.blobFromImage(cv2.resize(frame, (300, 300)), 0.007843, (300, 300), 127.5)
    
    net.setInput(blob)
    detections = net.forward()

    person_count = 0
    for i in range(detections.shape[2]):
        confidence = detections[0, 0, i, 2]
        if confidence > 0.5:
            idx = int(detections[0, 0, i, 1])
            # Check if the detected object is a person (class index 15 for MobileNetSSD)
            if idx == 15:
                person_count += 1
    return person_count
```

```python
# Process each folder and its videos
for root, dirs, files in os.walk(main_input_folder):
    # Iterate through each file in the current directory
    for file in tqdm(files):
        if file.endswith(".mp4"):
            video_path = os.path.join(root, file)
            print(f"Processing video: {video_path}")
            cap = cv2.VideoCapture(video_path)

            person_detected = False
            frame_skip = 6  # Process every 5th frame
            frame_count = 0
            # Iterate through video frames to count people
            while cap.isOpened():
                ret, frame = cap.read()
                if not ret:
                    break
                frame_count += 1
                if frame_count % frame_skip != 0:
                    continue
            
                person_count = count_people(frame)
                if person_count > 1:
                    person_detected = True
                    break

            cap.release()

            # If video has more than 1 person, move it to the output folder
            if person_detected:
                # Recreate subfolder structure in the output folder
                relative_path = os.path.relpath(root, main_input_folder)
                output_subfolder = os.path.join(output_folder, relative_path)
                os.makedirs(output_subfolder, exist_ok=True)

                # Move the video to the appropriate folder
                shutil.move(video_path, os.path.join(output_subfolder, file))
                print(f"Moved {file} to {output_subfolder}")

print("Processing complete.")
```

```python
import json
import csv
import os

# Path to the JSON file
json_file_path = r'F:\DFDC\JSON\metadata_19.json'

# Path to the output CSV file
csv_file_path = r'F:\DFDC\CSV\metadata_19.csv'

# Path to the folder containing the video files
video_folder_path = r'F:\DFDC\Data_2_people\Videos_19'

# Get a list of video files present in the folder
video_files_in_folder = set(os.listdir(video_folder_path))

# Open the JSON file and load the data
with open(json_file_path, 'r') as file:
    data = json.load(file)

# Open a CSV file for writing
with open(csv_file_path, 'w', newline='') as csvfile:
    # Create a CSV writer object
    csv_writer = csv.writer(csvfile)
    
    # Write the header row
    csv_writer.writerow(['Video Name', 'Label'])
    
    # Loop through the JSON data and write each video name and label to the CSV file
    for video_name, metadata in data.items():
        if video_name in video_files_in_folder:
            label = metadata.get('label', 'Unknown')  # Get label or default to 'Unknown'
            
            # Write the row to the CSV file
            csv_writer.writerow([video_name, label])
        else:
            print(f"Skipping {video_name} as it is not in the folder.")

print(f"Data successfully saved to {csv_file_path}")
```

```python
import cv2
import os
import numpy as np
import pandas as pd
from mtcnn import MTCNN
from deepface import DeepFace
from deep_sort_realtime.deepsort_tracker import DeepSort

# Initialize MTCNN for face detection
detector = MTCNN()

# Initialize DeepSORT for tracking
tracker = DeepSort(max_age=30)  # Adjust tracking settings

# Parent folder containing all video folders
PARENT_FOLDER = "F:/DFDC/Data_2_people/"  # Update with your actual path
OUTPUT_CSV = "F:/DFDC/face_tracking_results.csv"

# Initialize Data Storage
face_data = []

def process_video(video_path, folder_name, video_id):
    cap = cv2.VideoCapture(video_path)

    frame_number = 0
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break  # Stop if video ends

        frame_number += 1
        detections = []

        # Detect faces using MTCNN
        faces = detector.detect_faces(frame)
        for face in faces:
            x, y, w, h = face["box"]
            x, y, w, h = max(0, x), max(0, w), max(0, h), max(0, h)  # Ensure valid coordinates
            face_crop = frame[y:y+h, x:x+w]

            # Extract face embedding
            try:
                embedding = DeepFace.represent(face_crop, model_name="Facenet")[0]["embedding"]
            except:
                continue  # Skip if embedding fails

            # Store detection
            detections.append(([x, y, x + w, y + h], 0.99, "face", embedding))

        # Assign unique IDs using DeepSORT
        tracks = tracker.update_tracks(detections, frame=frame)

        for track in tracks:
            if not track.is_confirmed() or track.time_since_update > 1:
                continue  # Skip unconfirmed tracks

            track_id = track.track_id  # Unique ID for each person
            bbox = track.to_ltrb()  # Get bounding box (left, top, right, bottom)
            embedding = track.det_class  # Retrieve stored embedding

            # Store face tracking data
            face_data.append([
                folder_name, video_id, frame_number, track_id, bbox[0], bbox[1], bbox[2] - bbox[0], bbox[3] - bbox[1], embedding
            ])

        # Show results (optional)
        for track in tracks:
            x1, y1, x2, y2 = map(int, track.to_ltrb())
            cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
            cv2.putText(frame, f"ID {track.track_id}", (x1, y1 - 10),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 255, 0), 2)

    cap.release()
    cv2.destroyAllWindows()

# Iterate over all subfolders (Videos_00, Videos_01, ...)
for folder_name in os.listdir(PARENT_FOLDER):
    folder_path = os.path.join(PARENT_FOLDER, folder_name)
    
    if not os.path.isdir(folder_path):  
        continue  # Skip non-folder files
    
    print(f"Processing folder: {folder_name}...")
    
    # Iterate over all videos in the folder
    for filename in os.listdir(folder_path):
        if filename.endswith(".mp4") or filename.endswith(".avi"):
            video_path = os.path.join(folder_path, filename)
            print(f"   Processing video: {filename}...")
            process_video(video_path, folder_name, filename)

# Convert data to DataFrame and save as CSV
df = pd.DataFrame(face_data, columns=["folder_name", "video_id", "frame_number", "face_id", "bbox_x", "bbox_y", "bbox_w", "bbox_h", "embedding"])
df.to_csv(OUTPUT_CSV, index=False)

print(f"✅ Face tracking data saved to {OUTPUT_CSV}")
```
